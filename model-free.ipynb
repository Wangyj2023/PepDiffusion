```python
from collections import Counter
import pandas as pd

with open(f"data/pos_data", 'r') as file:
    sequences = file.read().splitlines()
aa_distribution = Counter()
for seq in sequences:
    aa_distribution.update(seq)
position_distribution = {}
max_length = max(len(seq) for seq in sequences)
for i in range(max_length):
    position_distribution[i] = Counter()
for seq in sequences:
    for i, aa in enumerate(seq):
        position_distribution[i][aa] += 1
length_distribution = Counter(len(seq) for seq in sequences)
aa_distribution_df = pd.DataFrame.from_dict(aa_distribution, orient='index', columns=['Frequency']).sort_values(by='Frequency', ascending=False)
position_distribution_df = {pos: pd.DataFrame.from_dict(dist, orient='index', columns=['Frequency']).sort_values(by='Frequency', ascending=False) for pos, dist in position_distribution.items()}
length_distribution_df = pd.DataFrame.from_dict(length_distribution, orient='index', columns=['Frequency']).sort_index()
import random

def generate_peptide_by_overall_distribution(aa_dist, length_dist):
    lengths, length_probs = zip(*length_dist.items())
    total_length = sum(length_dist.values())
    length_probs = [freq / total_length for freq in length_probs]
    selected_length = random.choices(lengths, weights=length_probs, k=1)[0]
    aas, aa_probs = zip(*aa_dist.items())
    total_aa_freq = sum(aa_dist.values())
    aa_probs = [freq / total_aa_freq for freq in aa_probs]
    peptide = ''.join(random.choices(aas, weights=aa_probs, k=selected_length))
    return peptide

def generate_peptide_by_position_distribution(pos_dist, length_dist):
    lengths, length_probs = zip(*length_dist.items())
    total_length = sum(length_dist.values())
    length_probs = [freq / total_length for freq in length_probs]
    selected_length = random.choices(lengths, weights=length_probs, k=1)[0]
    peptide = ''
    for i in range(selected_length):
        aas, aa_probs = zip(*pos_dist[i].items())
        total_aa_freq = sum(pos_dist[i].values())
        aa_probs = [freq / total_aa_freq for freq in aa_probs]
        peptide += random.choices(aas, weights=aa_probs, k=1)[0]
    return peptide

with open(f"generate_peptide_by_overall_distribution","w") as wf:
    for _ in range(5120):
        wf.write(f"{generate_peptide_by_overall_distribution(aa_distribution, length_distribution)}\n")
with open(f"generate_peptide_by_position_distribution","w") as wf:
    for _ in range(5120):
        wf.write(f"{generate_peptide_by_position_distribution(position_distribution, length_distribution)}\n")

def random_generate(length_dist):
    lengths, length_probs = zip(*length_dist.items())
    total_length = sum(length_dist.values())
    length_probs = [freq / total_length for freq in length_probs]
    selected_length = random.choices(lengths, weights=length_probs, k=1)[0]
    amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
    return ''.join(random.choices(amino_acids, k=selected_length))

with open(f"random_sample","w") as wf:
    for _ in range(22000):
        wf.write(f"{random_generate(length_distribution)}\n")

import random


file_path = 'data/VAE_Train'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
if len(lines) < 220:
    print("文件中的行数少于220行。")
else:
    selected_lines = random.sample(lines, 22000)
    with open('uniprot_sample', 'w', encoding='utf-8') as output_file:
        for line in selected_lines:
            output_file.write(line)
```
